```python
class Value:
  def __init__(self, data: float, parrents_set: set = None, operation: str = None):
    self.data = data
    self._prev = parrents_set
    self._op = operation

  def __repr__(self) -> str:
    return f"Value(data={self.data})"


  def __add__(self, other) -> Value:
    if not isinstance(other, Value):
      other = Value(other)
    return Value(self.data + other.data, parrents_set={self, other}, operation='+')

  def __mul__(self, other) -> Value:
    if not isinstance(other, Value):
      other = Value(other)
    return Value(self.data * other.data, parrents_set={self, other}, operation='*')

  def trace(object: Value) -> tuple[set, set]:
    def trace_help(node: Value):
      if node not in nodes:
        nodes.add(node)
        if node._prev:
          for parent in node._prev:
            edges.add((parent, node))
            trace_help(parent)

    nodes, edges = set(), set()
    trace_help(object)
    return nodes, edges

  def draw_dot(root: Value) -> graphviz.Digraph:
      dot = graphviz.Digraph(filename='01_result', format='svg', graph_attr={
                            'rankdir': 'LR'})  # LR = left to right

      nodes, edges = trace(root)
      for n in nodes:
          uid = str(id(n))
          # for any value in the graph, create a rectangular ('record') node
          dot.node(name=uid, label=f'{{ data: {n.data} }}', shape='record')
          if n._op:
              # if this value is a result of some operation, create an "op" node for the operation
              dot.node(name=uid + n._op, label=n._op)
              # and connect this node to the node of the operation
              dot.edge(uid + n._op, uid)

      for n1, n2 in edges:
          # connect n1 to the "op" node of n2
          dot.edge(str(id(n1)), str(id(n2)) + n2._op)

      return dot


x = Value(2.0)
y = Value(-3.0)
z = Value(10.0)
result = x * y + z

# This will create a new directory and store the output file there.
# With "view=True" it'll automatically display the saved file.
# draw_dot(result).render(directory='./graphviz_output', view=True)


dot = draw_dot(result)
dot.render(directory='/content/graphviz_output', format='png')
```

```python
class Value:
  def __init__(self, data: float, parrents_set: set = None, operation: str = None):
    self.data = data
    self._prev = parrents_set
    self._op = operation
    self.grad = 0.0

  def __repr__(self) -> str:
    return f"Value(data={self.data})"


  def __add__(self, other) -> Value:
    if not isinstance(other, Value):
      other = Value(other)
    return Value(self.data + other.data, parrents_set={self, other}, operation='+')

  def __mul__(self, other) -> Value:
    if not isinstance(other, Value):
      other = Value(other)
    return Value(self.data * other.data, parrents_set={self, other}, operation='*')

def trace(object: Value) -> tuple[set, set]:
  def trace_help(node: Value):
    if node not in nodes:
      nodes.add(node)
      if node._prev:
        for parent in node._prev:
          edges.add((parent, node))
          trace_help(parent)

  nodes, edges = set(), set()
  trace_help(object)
  return nodes, edges

def draw_dot(root: Value) -> graphviz.Digraph:
  dot = graphviz.Digraph(filename='Task10', format='svg', graph_attr={'rankdir': 'LR'})
  nodes, edges = trace(root)
  for n in nodes:
    uid = str(id(n))

    temp_l = chr(int(uid) % 26 + ord('a'))

    dot.node(name=uid, label=f'{{ {temp_l} | data: {n.data} | grad: {n.grad} }}', shape='record')
    if n._op:
      dot.node(name=uid + n._op, label=n._op)
      dot.edge(uid + n._op, uid)

  for n1, n2 in edges:
    dot.edge(str(id(n1)), str(id(n2)) + n2._op)

  return dot


a = Value(2.0)
b = Value(-3.0)
c = a * b
d = Value(10.0)
e = c + d
f = Value(-2.0)
L = e * f

# Start from the final child (the last node in topological order).
# + => copy gradient to parents:
#     parent1.grad = current.grad
#     parent2.grad = current.grad
# * => multiply value of other parent with current gradient:
#     parent1.grad = parent2.value * current.grad
#     parent2.grad = parent1.value * current.grad

L.grad = 1.0

f.grad = L.grad * e.data  # 4.0
e.grad = L.grad * f.data  #-2.0
d.grad = e.grad # -2.0
c.grad = e.grad # -2.0
b.grad = c.grad * a.data  # 6.0
a.grad = c.grad * b.data  # -4.0

def manual_der(node, epsilon=0.001):
  a_pert = Value(a.data)
  b_pert = Value(b.data)
  c_pert = Value(c.data)
  d_pert = Value(d.data)
  e_pert = Value(e.data)
  f_pert = Value(f.data)
  L_pert = Value(L.data)

  if node is a:
    a_pert.data += epsilon
    c_pert = a_pert * b_pert
    e_pert = c_pert + d_pert
    L_pert = e_pert * f_pert
  elif node is b:
    b_pert.data += epsilon
    c_pert = a_pert * b_pert
    e_pert = c_pert + d_pert
    L_pert = e_pert * f_pert
  elif node is c:
    c_pert.data += epsilon
    e_pert = c_pert + d_pert
    L_pert = e_pert * f_pert
  elif node is d:
    d_pert.data += epsilon
    e_pert = c_pert + d_pert
    L_pert = e_pert * f_pert
  elif node is e:
    e_pert.data += epsilon
    L_pert = e_pert * f_pert
  elif node is f:
    f_pert.data += epsilon
    L_pert = e_pert * f_pert
  elif node is L:
    L_pert += epsilon

  numerical_derivative = (L_pert.data - L.data) / epsilon
  return numerical_derivative

print(f"{a.grad=}, {manual_der(a)=}")
print(f"{b.grad=}, {manual_der(b)=}")
print(f"{c.grad=}, {manual_der(c)=}")
print(f"{d.grad=}, {manual_der(d)=}")
print(f"{e.grad=}, {manual_der(e)=}")
print(f"{f.grad=}, {manual_der(f)=}")

draw_dot(L).render(directory='/content/graphviz_output', format='png', filename='Task10')
```

```python
class Value:
  def __init__(self, data: float, parrents_set: set = set(), operation: str = None, label = None, backwards = lambda _ : None):
    self.data = data
    self.grad = 0.0
    self.label = label
    self._prev = parrents_set
    self._op = operation
    self._backward = backwards
    if label is None:
      self.label = '_' + chr(int(id(self)) % 26 + ord('a'))

  def __repr__(self) -> str:
    return f"{self.label} Value(data={self.data})"

  def __add__(self, other) -> 'Value':
    def _add_backward(next_grad):
      self.grad = next_grad
      other.grad = next_grad

    if not isinstance(other, Value):
      other = Value(other)
    return Value(self.data + other.data, parrents_set={self, other}, operation='+', backwards=_add_backward)

  def __mul__(self, other) -> 'Value':
    if not isinstance(other, Value):
      other = Value(other)
    return Value(self.data * other.data, parrents_set={self, other}, operation='*')

  def tahn(self) -> 'Value':
    left_side = (math.exp(self.data) - math.exp(self.data*(-1)))
    right_side = (math.exp(self.data) + math.exp(self.data*(-1)))
    return Value(left_side / right_side, parrents_set={self}, operation='tanh')



def top_sort(node) -> list:
  visited = set()
  order = []

  def topo_help(v):
    if v not in visited:
      visited.add(v)
      for parent in v._prev:
        topo_help(parent)
      order.append(v)


  topo_help(node)
  return order

def backward(start_node: 'Value'):
  start_node.grad = 1.0
  order = top_sort(start_node)
  order.reverse()
  for node in order:
    node._backward(node.grad)


x = Value(2.0, label='x')
y = Value(-3.0, label='y')
z = Value(7.0, label='z')
a = x + y
a.label = 'a'
L = a + z
L.label = 'L'

backward(L)

draw_dot(L).render(directory='/content/graphviz_output', format='png', filename='Task15')
```
