# Task 36

```python
import random
import numpy as np
import matplotlib.pyplot as plt

random.seed(123)

def random_walk():
  step = 0
  floors = []
  for i in range(100):
    if random.random() <= 0.005:
      step = 0
    else:
      dice = random.randint(1, 6)
      if dice == 1 or dice == 2:
        step -= 1
      elif dice in [3, 4, 5]:
        step += 1
      else:
        dice2 = random.randint(1, 6)
        step += dice2
      if step < 0:
        step = 0
    floors.append(step)
  return floors


all_walks = []

for _ in range(20):
  all_walks.append(random_walk())

walks_array = np.array(all_walks)

plt.plot(walks_array.T)
plt.title('Random walks')
plt.ylabel('Floor Number')
plt.show()
```
