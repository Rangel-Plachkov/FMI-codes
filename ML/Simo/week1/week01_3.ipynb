```python
np.random.seed(123)

step = 0
floors = []

for i in range(100):
    dice = np.random.randint(1, 7)

    if dice == 1 or dice == 2:
        step -= 1
    elif dice in [3, 4, 5]:
        step += 1
    else:
        dice2 = np.random.randint(1, 7)
        step += dice2

    if step < 0:
        step = 0

    floors.append(step)

print(floors)
```

# Task 34

```python
np.random.seed(123)

def random_walk():
    step = 0
    floors = []
    for i in range(100):
        dice = np.random.randint(1, 7)
        if dice == 1 or dice == 2:
            step -= 1
        elif dice in [3, 4, 5]:
            step += 1
        else:
            dice2 = np.random.randint(1, 7)
            step += dice2
        if step < 0:
            step = 0
        floors.append(step)
    return floors

all_walks = []

for _ in range(5):
    all_walks.append(random_walk())

print(all_walks)
```

# Task 35

```python
walks_array = np.array(all_walks)


plt.plot(walks_array.T)
plt.title('Random walks')
plt.xlabel('Throw')
plt.show()
```

# Task 36

```python
np.random.seed(123)

def random_walk():
    step = 0
    floors = []
    for i in range(100):
        if np.random.random() <= 0.005:
            step = 0
        else:
            dice = np.random.randint(1, 7)
            if dice == 1 or dice == 2:
                step -= 1
            elif dice in [3, 4, 5]:
                step += 1
            else:
                dice2 = np.random.randint(1, 7)
                step += dice2
            if step < 0:
                step = 0
        floors.append(step)
    return floors

all_walks = []

for _ in range(20):
    all_walks.append(random_walk())

walks_array = np.array(all_walks)

plt.plot(walks_array.T)
plt.title('Random walks')
plt.ylabel('Floor Number')
plt.show()
```
